```python
results = pd.read_csv("results.csv")
print(results[:5])
```

```python
counts = [0] * 20
r_counts = [0] * 20
loc = 0 
for i in range(len(results)):
    start = results["grid"][i]
    loc = results["positionText"][i]
    if start < 20:
        counts[start - 1] += 1
        if loc == "R":
            r_counts[start - 1] += 1


print(r_counts)
print(counts)

per = []
for i in range(20):
    per.append(r_counts[i] / counts[i])

print(per)
```

```python
positions = np.linspace(1, 19, 19)
plt.plot(positions, per[:19])
plt.xticks(positions)
plt.xlabel('Starting Position')
plt.ylabel('Proportion of Retirements')
plt.title('Proportion of Retirements by Starting Position')
plt.grid()
plt.show()
print(per)
```
